```python
import pandas
```

```python
initial_revenue = 222
start_year = '2022'
end_year = '2032'
revenue_growth_rate = 1.18
ebitda_margin = 0.45
Taxes= 0.21 
Capital_Expenditures= 0.03
Working_Capital_Increase= 0.06
EBITDA_Exit_Multiple= 32.70
Discount_Factor= 0.94
```

```python
dates= pandas.date_range(start=start_year,end=end_year,freq='Y')
dates
```

```python
df = pandas.DataFrame(index=dates)
terminal_value_df = pandas.DataFrame()
df
```

```python
initial_revenue
rev= [ initial_revenue* 1.18]


for x in range(1, len(dates)):
  rev.append(rev[x-1]*revenue_growth_rate)

df['Revenue'] = rev
df
```

```python
df["EBITDA"] =df ["Revenue"] * ebitda_margin
df
```

```python
df['Taxes']=-df['EBITDA']*Taxes
df
```

```python
df["Capital Expenditure"] = -df ["Revenue"] * 0.03
df
```

```python
df['Working Capital']=-df['Revenue']* Working_Capital_Increase
df
```

```python
df['Cash flow']=df['EBITDA'] + df['Taxes'] + df['Capital Expenditure'] + df['Working Capital']
df
```

```python
df['Discount_Factor']= [Discount_Factor** i for i in range(1,len(dates)+1)]
df
```

```python
df['Present Value of Future Cash Flow']= df['Cash flow']* df['Discount_Factor']
df
```

```python
Revenue_2032= pandas.DataFrame(index=['2032'])
Revenue_2032= df['Revenue'].iloc[-1]*revenue_growth_rate
Revenue_2032.round()
```

```python
ebitda_2032= pandas.DataFrame(index=['2032'])
ebitda_2032= Revenue_2032*ebitda_margin
ebitda_2032.round()
```

```python
terminal_value=ebitda_2032 * EBITDA_Exit_Multiple 
pv=df['Present Value of Future Cash Flow'].sum().round(2)
Terminal_Value= terminal_value+pv
Terminal_Value.round()
```

```python
print(f""" Our projected Present Value of Projected Future Cash Flows are:
 ${df['Present Value of Future Cash Flow'].sum().round(2)}.

We are providing revenue guidance in 2032 to be:
 ${round(df.iloc[-1]['Revenue']*revenue_growth_rate)}

With an estimated EBITDA around:
 ${round(ebitda_2032)}

The terminal value a this time given a multiple of 32.70 and the above EBITDA for 2030, would equal:
 ${round(Terminal_Value)}. 
 
     """)
```
